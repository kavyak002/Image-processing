```python
from skimage import color
from skimage import io
img=io.imread('flower1.jpg')
imgGray=color.rgb2gray(img)
io.imshow(imgGray)
```

```python
from PIL import Image
img=Image.open('flower1.jpg')
imgGray=img.convert('L')
imgGray.show()
```

```python
import cv2
img_gray=cv2.imread("flower1.jpg",0)
cv2.imshow('Grayscale conversion OpenCV',img_gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import glob
import matplotlib.pyplot as plt

path="C:/Users/User/Image Processing23/images/*.*"

images=[cv2.imread(image) for image in glob.glob(path)]
fig=plt.figure()
for i in range(len(images)):
    plt.subplot(5,5,i+1)
    plt.imshow(images[i])
    plt.show()
```

```python
import cv2
import glob

path=("C:/Users/User/Image Processing23/images/*.*")
for file in glob.glob(path):
    image_read=cv2.imread(file)
    c=cv2.cvtColor(image_read,cv2.COLOR_BGR2RGB)
    cv2.imshow('Color image',c)
    k=cv2.waitKey(2000)
    cv2.destroyAllWindows()
```

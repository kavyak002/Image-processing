```python
#Affine transformation
import cv2
import numpy as np
from matplotlib import pyplot as plt
 
img = cv2.imread('bridge.jpg')
rows, cols, ch = img.shape
 
pts1 = np.float32([[50, 50],
                   [200, 50],
                   [50, 200]])
 
pts2 = np.float32([[10, 100],
                   [200, 50],
                   [100, 250]])
 
M = cv2.getAffineTransform(pts1, pts2)
dst = cv2.warpAffine(img, M, (cols, rows))
 
plt.subplot(121)
plt.imshow(img)
plt.title('Input')
 
plt.subplot(122)
plt.imshow(dst)
plt.title('Output')
 
plt.show()
         
cv2.destroyAllWindows()
```

```python
#Removing Logo’s
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from skimage.restoration import inpaint
from skimage.transform import resize
from skimage import color
image_with_logo=plt.imread('logo.jpg')
mask=np.zeros(image_with_logo.shape[:-1])
mask[25:152,25:190]=1
image_logo_removed=inpaint.inpaint_biharmonic(image_with_logo,
mask,multichannel=True)
plt.title('image with logo')
plt.imshow(image_with_logo)
plt.show()
plt.title('image without logo')
plt.imshow(image_logo_removed)
plt.show()
```

```python
# Importing the OpenCV, Numpy and Mat libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt


img = cv2.imread('jambo.jpg')
plt.imshow(img)
plt.show()



box_blur_ker = np.array([[0.1111111, 0.1111111, 0.1111111],
                         [0.1111111, 0.1111111, 0.1111111],
                         [0.1111111, 0.1111111, 0.1111111]])


Box_blur = cv2.filter2D(src=img, ddepth=-1, kernel=box_blur_ker)


plt.imshow(Box_blur)
plt.show()
```

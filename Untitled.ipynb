```python
#Feature extraction from Image

#Importing the important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from skimage.io import imread,imshow

#Loading the image
image1=imread('jambo.jpg')
imshow(image1);
```

```python
#Converting te original image to gray
image2=imread('jambo.jpg',as_gray=True)
imshow(image2);
```

```python
#Analyzing both the images
print(image1.shape)
print(image2.shape)
```

```python
#size of the images
print(image1.size)
print(image2.size)
```

```python
#Feature extraction
#i.Pixel Features
pixel_features1=np.reshape(image2,(168*300))
pixel_features1.shape,pixel_features1
```

```python
#Pixel features for the colored image
pixel_features2=np.reshape(image1,(168*300*3))
pixel_features2.shape,pixel_features2
```

```python
#Calculating the variance of all pixels for each band na an image using the Pillow library.
from PIL import Image,ImageStat
im=Image.open("jambo.jpg")
stat=ImageStat.Stat(im)
print(stat.var)
```
